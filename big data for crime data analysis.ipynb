```python
import numpy as np

# Assuming 'violent_crimes' is a column in your dataframe and 'violent_crimes_mean' is the mean value of violent crimes.
violent_crimes_mean = df['violent_crimes'].mean()  # Calculate the mean of violent crimes if not already defined.

# Add a column with the mean of violent crimes
df['mean_violent_crimes'] = violent_crimes_mean

# Categorize rows based on whether violent crimes are above or below the mean
df['violent_crime_occurrence'] = np.where(df['violent_crimes'] >= df['mean_violent_crimes'], 1, 0)

# Group by violent crime occurrence and calculate the mean of other numerical columns
result = df.groupby('violent_crime_occurrence').mean()

# Display the result
print(result)
```

```python
import numpy as np

# Calculate the mean of violent crimes per population
violent_crimes_mean = df['ViolentCrimesPerPop'].mean()

# Add a column with the mean of violent crimes per population
df['mean_violent_crimes'] = violent_crimes_mean

# Categorize rows based on whether violent crimes per population are above or below the mean
df['violent_crime_occurrence'] = np.where(df['ViolentCrimesPerPop'] >= df['mean_violent_crimes'], 1, 0)

# Group by violent crime occurrence and calculate the mean of other numerical columns
result = df.groupby('violent_crime_occurrence').mean()

# Display the result
print(result)
```

```python
# Convert 'ViolentCrimesPerPop' to numeric, forcing errors to NaN
df['ViolentCrimesPerPop'] = pd.to_numeric(df['ViolentCrimesPerPop'], errors='coerce')

# Optionally, drop rows with NaN values in the 'ViolentCrimesPerPop' column
df = df.dropna(subset=['ViolentCrimesPerPop'])

# Now calculate the mean again
violent_crimes_mean = df['ViolentCrimesPerPop'].mean()

# Add a column with the mean of violent crimes per population
df['mean_violent_crimes'] = violent_crimes_mean

# Categorize rows based on whether violent crimes per population are above or below the mean
df['violent_crime_occurrence'] = np.where(df['ViolentCrimesPerPop'] >= df['mean_violent_crimes'], 1, 0)

# Group by violent crime occurrence and calculate the mean of other numerical columns
result = df.groupby('violent_crime_occurrence').mean()

# Display the result
print(result)
```
